```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
file_path = "lex.csv"
df = pd.read_csv(file_path)
df
```

```python
df.set_index("country", inplace=True)
df = df.transpose()  # Years become the index
 

```

```python
# Convert index to numeric (years)
df.index = pd.to_numeric(df.index, errors='coerce')

```

```python
# Calculate the average life expectancy per year
df["Average Life Expectancy"] = df.mean(axis=1, skipna=True)
```

```python
# Plot the chart
plt.figure(figsize=(10, 6))
plt.plot(df["Average Life Expectancy"], df.index, marker='o', linestyle='-', color='b', label="Avg Life Expectancy")
```

```python
# Labels and title
plt.xlabel("Average Life Expectancy")
plt.ylabel("Year")
plt.title("Change in Life Expectancy Over the Years")
plt.legend()
plt.grid(True)
```

```python
# Show theÂ plot
plt.show()
```
